```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Loading Data
```

```python
credit_card_data=pd.read_csv(r'C:\Users\dell\Desktop\creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.info()
```

```python
credit_card_data['Class'].value_counts()
```

```python
# Under Sampling of data as data is unbalanced
```

```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
legit.describe()
```

```python
fraud.describe()
```

```python
legit_Data=legit.sample(n=492)
legit_Data.shape
```

```python
new_data=pd.concat([legit_Data,fraud],axis=0)
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
#Splitting of Data
```

```python
X=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
print(X)
```

```python
print(y)
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=2,stratify=y)
```

```python
#Applying Logistic Regression Model for prediction
```

```python
model=LogisticRegression()
```

```python
model.fit(Xtrain,ytrain)
```

```python
y_pred=model.predict(Xtest)
print(y_pred)
```

```python
# Accuracy Score of Training Set and Test Set
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_train=model.predict(Xtrain)
print(y_pred_train)
```

```python
accuracy_score(y_pred_train,ytrain)
```

```python
accuracy_score(y_pred,ytest)
```
